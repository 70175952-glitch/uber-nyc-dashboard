```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Dark theme
plt.rcParams.update({'figure.facecolor':'#0b0f1a','axes.facecolor':'#1a2235',
                     'text.color':'#e2e8f0','axes.labelcolor':'#64748b',
                     'xtick.color':'#64748b','ytick.color':'#64748b',
                     'axes.edgecolor':'#1e2d45','grid.color':'#1e2d45'})
print('Libraries loaded ✓')
```

```python
# Load dataset
df = pd.read_csv('../data/uber-raw-data-apr14.csv')
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

# Feature engineering
df['Hour']      = df['Date/Time'].dt.hour
df['DayOfWeek'] = df['Date/Time'].dt.day_name()
df['Month']     = df['Date/Time'].dt.month
df['DayNum']    = df['Date/Time'].dt.dayofweek
df['WeekOfYear']= df['Date/Time'].dt.isocalendar().week.astype(int)

print(f'Shape: {df.shape}')
df.head()
```

```python
# Basic info
print(df.info())
print('\nMissing values:')
print(df.isnull().sum())
print('\nBasic statistics:')
df.describe()
```

```python
# Hourly trend
fig, ax = plt.subplots(figsize=(12,4))
hourly = df.groupby('Hour').size()
ax.plot(hourly.index, hourly.values, color='#00c6ff', linewidth=2, marker='o', ms=4)
ax.fill_between(hourly.index, hourly.values, alpha=0.2, color='#00c6ff')
ax.set_title('Uber Pickups by Hour of Day', fontsize=14, fontweight='bold')
ax.set_xlabel('Hour'); ax.set_ylabel('Pickups')
plt.tight_layout(); plt.show()
```

```python
# Correlation heatmap
num_cols = ['Hour','Lat','Lon','DayNum','Month','WeekOfYear']
corr = df[num_cols].corr()
fig, ax = plt.subplots(figsize=(7,5))
sns.heatmap(corr, ax=ax, annot=True, fmt='.2f',
            cmap=sns.diverging_palette(220,10,as_cmap=True),
            linewidths=0.5, annot_kws={'size':9})
ax.set_title('Feature Correlation Heatmap', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.show()
```
